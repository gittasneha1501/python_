```python
import seaborn as sns
import numpy as np
ds=sns.get_dataset_names()
print(ds)
print(len(ds))
```

```python
#lineplot
data=sns.load_dataset('diamonds')
print(data)
sns.lineplot(x='carat',y='price',data=data)
```

```python
#barplot
data=sns.load_dataset('attention')
print(data)
sns.barplot(x='score',y='solutions',data=data)
```

```python
#countplot
data=sns.load_dataset('attention')
print(data)
sns.countplot(x='solutions',data=data)
```

```python
#boxplot
data=sns.load_dataset('anscombe')
print(data)
sns.boxplot(x='x',y='y',data=data)
```

```python
#violinplot
data=sns.load_dataset('planets')
print(data)
sns.violinplot(x='method',y='mass',data=data)
```

```python

#stripplot
data=sns.load_dataset('planets')
print(data)
sns.stripplot(x='mass',y='year',data=data)
```

```python
#histplot
data=sns.load_dataset('planets')
print(data)
sns.histplot(y='year',data=data,kde=True)
```

```python
#pairplot
data=sns.load_dataset('attention')
print(data)
sns.pairplot(hue='score',data=data)
```

```python
data=np.random.randint(low=2,high=80,size=(8,8))
cmap='tab20'
center=0
annot=True
hm=sns.heatmap(data=data,annot=annot,cmap=cmap,center=center)
```

```python
#catplot
data=sns.load_dataset('exercise')
print(data)
sns.catplot(x='id',y='pulse',hue='time',data=data)
```
