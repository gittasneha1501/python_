#types

```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
#line plot
x=[1,2,3,4,5]
y=[6,7,8,9,10]
plt.plot(x,y)
plt.show()
```

```python
plt.bar(x,y)   #bar plot
```

```python
#histogram plot
x=[1,2,3,4,5]
y=[6,7,8,9,10]
plt.hist(x,y)
plt.show()
```

```python
#scatter plot
plt.scatter(x,y)
```

```python
#pie chart
groceries=['rice','dal','oil','vegetables']
size=[45,10,15,30]
plt.pie(size,labels=groceries,startangle=90)
plt.title('expenditure on groceries')
```

```python
#area chart
x=[1,2,3,4]
y1,y2=[5,6,7,8],[2,4,6,8]
plt.fill_between(x,y1,y2,color='orange',alpha=0.6,label='area')
plt.plot(x,y1,label='line1',marker='*')
plt.plot(x,y2,label='line2',marker='*')
plt.xlabel('x_axis'),plt.ylabel('y_axis'),plt.title('chart of area')
```

#introduction to figures


```python
draw=plt.figure(figsize=(5,8),facecolor='lightblue',layout='constrained')
draw.suptitle('drawing')
x=draw.add_subplot()
x.set_title('figure',loc='center',fontstyle='italic',fontsize='small')
```

#subplot[]function

```python

fig,a= plt.subplots(2,2)
x=np.arange(1,10)
a[0][0].plot(x,x*x)
a[0][0].set_title('square')
a[0][1].plot(x,np.sqrt(x))
a[0][1].set_title('square_root')
a[1][0].plot(x,np.exp(x))
a[1][0].set_title('exponential')
a[1][1].plot(x,np.log10(x))
a[1][1].set_title('log')
```
